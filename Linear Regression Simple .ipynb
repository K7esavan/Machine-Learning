# Linear Regression - Single Variable

    Here we have only 2 variables and one is dependent and the other is independent 
    
   **y=mx+c**

Lets take example price and area of house 

Here our target variable is price which is dependent on area of the house

```python
# required headers
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
home_df = pd.DataFrame({
    
    'Area' : [  2600,   3000,   3200,   3600,   4000],
    'Price': [550000, 565000, 610000, 680000, 725000]
})
```

```python
%matplotlib inline
plt.xlabel(' Area (sq.feet) ')
plt.ylabel(' Price( US $))  ')
plt.scatter(home_df.Area, home_df.Price, color='red', marker='+')
```

```python
# Now lets create the simple regression model

reg = linear_model.LinearRegression()
reg.fit(home_df[['Area']], home_df.Price)
```

```python
# Here goes the predictions 

print (reg.predict([[5012]]))
print (reg.coef_)
print (reg.intercept_)
```

```python
%matplotlib inline
plt.xlabel(' Area (sq.feet) ')
plt.ylabel(' Price( US $))  ')
plt.scatter(home_df.Area, home_df.Price, color='red', marker='+')
plt.plot(home_df.Area, reg.predict(home_df[['Area']]), color='blue')
```
